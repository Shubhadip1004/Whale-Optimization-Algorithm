Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Importing Whale Optimization Algorithm and Benchmark functions files

```python
import Whale_Optimization_Algorithm as woa
import Benchmark_Functions as functions
```

```python
itr = 500
n = 30
xlst = [i for i in range(1, itr+1)]
```

Sphere Function 

```python
(lb, ub) = ([-5.12], [5.12])
dim = 30
pos_whale = np.random.uniform(lb, ub, [n, dim]) 
data = woa.run_woa(pos_whale, functions.sphere, lb, ub, dim, itr, n)
ylst = data[:itr]
best_pos = list(map(float,data[(-(dim + 1)):-1]))    
best_fx = data[-1]
print("Best fitness value obtained for Sphere function is:", best_fx)
print("Best position obtained for Sphere function is:", best_pos)
plt.plot(xlst, ylst)
plt.title("Sphere Function")
plt.xlabel("Iterations")
plt.ylabel("Fitness value")
plt.show()
```
