Implementing Whale Optimization Algortihm to optimize Engineering Design Problems

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Importing Whale Optimization Algorithm and Engineering Design Problems

```python
import Whale_Optimization_Algorithm as woa
import Engineering_Design as functions
```

```python
itr = 500
n = 30
xlst = [i for i in range(1, itr+1)]
```

Compression/Tension Spring Design

```python
def initialise_pos_PVD(n, dim):
    pos = []
    for _ in range(n):
        temp = []
        for i in range(dim):
            if i == 0:
                x = random.uniform(0.05, 2)
            elif i == 1:
                x = random.uniform(0.25, 1.3)
            else:
                x = random.uniform(2, 15)
            temp.append(x)
        pos.append(temp)
    return np.array(pos)
```

```python
lb = [0.05, 0.25, 2]
ub = [2, 1.3, 15]
dim = 3
pos_whale = initialise_pos_PVD(n, dim)
data = woa.run_woa(pos_whale, functions.CSD, lb, ub, dim, itr, n)
ylst = data[:itr]
best_pos = list(map(float,data[(-(dim + 1)):-1]))    
best_fx = data[-1]
print("Best fitness value obtained for Compression Spring Design function is:", best_fx)
print("Best position obtained for Compression Spring Design function is:", best_pos)
plt.plot(xlst, ylst)
plt.title("Compression Spring Design Function")
plt.xlabel("Iterations")
plt.ylabel("Fitness value")
plt.show()
```

Welded Beam Design

```python
def initialise_pos_WBD(n, dim):
    pos = []
    for _ in range(n):
        temp = []
        for i in range(dim):
            if i == 0:
                x = random.uniform(0.1, 2)
            elif i == 1:
                x = random.uniform(0.1, 10)
            elif i == 2:
                x = random.uniform(0.1, 10)
            else:
                x = random.uniform(0.1, 2)
            temp.append(x)
        pos.append(temp)
    return np.array(pos)
```

Speed Reducer Problem

```python
def initialise_pos_SRP(n, dim):
    pos = []
    for _ in range(n):
        temp = []
        for i in range(dim):
            if i == 0:
                x = random.uniform(2.6, 3.6)
            elif i == 1:
                x = random.uniform(0.7, 0.8)
            elif x == 2:
                x = random.uniform(17, 28)
            elif x == 3:
                x = random.uniform(7.3, 8.3)
            elif x == 4:
                x = random.uniform(7.8, 8.3)
            elif x == 5:
                x = random.uniform(2.9, 3.9)
            else:
                x = random.uniform(5, 5.5)
            temp.append(x)
        pos.append(temp)
    return np.array(pos)
```
